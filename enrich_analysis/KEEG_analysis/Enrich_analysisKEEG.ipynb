View of the dataframe sorted by GeneRatio

```r
# Load object saved from KEGG enrichment
simp_resultsKEGG <- readRDS("results/enrich_resultsKEGG.rds")

# Convert to tibble
df_results <- as_tibble(simp_resultsKEGG@result)

# Filter significant terms and sort by descending GeneRatio
df_sig_ordered <- df_results %>%
  filter(pvalue < 0.05) %>%
  mutate(GeneRatio_num = as.numeric(sapply(strsplit(GeneRatio, "/"), function(x) x[1])) /
                    as.numeric(sapply(strsplit(GeneRatio, "/"), function(x) x[2]))) %>%
  arrange(desc(GeneRatio_num)) %>%
  select(-GeneRatio_num)

# View sorted results
tibble(df_sig_ordered)
```

Simplified visualization by enriched term to generate plots on the KEGG website

```r
library(AnnotationDbi)
library(org.Hs.eg.db)
library(dplyr)

# Filter terms with p-value < 0.05
sig_terms <- simp_resultsKEGG@result %>% 
  filter(pvalue < 0.05)

# Get all unique Entrez IDs
all_entrez <- sig_terms$geneID %>%
  strsplit("/") %>%
  unlist() %>%
  unique()

# Map Entrez IDs to gene symbols
map_tbl <- AnnotationDbi::select(
  org.Hs.eg.db,
  keys     = all_entrez,
  keytype  = "ENTREZID",
  columns  = c("SYMBOL")
) %>%
  distinct(ENTREZID, .keep_all = TRUE)

# Show the terms and their p-values
for (i in seq_len(nrow(sig_terms))) {
  
  ids <- unlist(strsplit(sig_terms$geneID[i], "/"))
  
  # Get gene symbols
  gene_symbols <- map_tbl %>%
    filter(ENTREZID %in% ids) %>%
    pull(SYMBOL)
  
  cat("Termo:", sig_terms$Description[i], "\n")
  cat("ID:", sig_terms$ID[i], "\n")
  cat("p-value:", sig_terms$pvalue[i], "\n")
  cat("Genes (Entrez IDs):", gsub("/", " ", sig_terms$geneID[i]), "\n")
  cat("Gene Symbols:", paste(gene_symbols, collapse = " "), "\n\n")
}
```
